```python

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

# input is a string, but we need a number which we will use as index to retrieve an item from a list
my_choice = int(input('what do you choose? type 0 for Rock, 1 for Paper and 2 for Scissors '))

# we have to check for invalid number as well
if my_choice >= 3 or my_choice < 0:
    print('you typed an invalid number, you lost')
    
    # once the conditin is true, it executes the rest of our code 
else:
    # store variable in a list, so that we can have access to them through an index
    choices = [rock, paper, scissors]

    # we use the input's result as an index to access the list choices --> example, choices[0] is a rock
    choices[my_choice]

    import random

    # generate a random number for computer and use it as index in choices[]
    computer = random.randint(0, 2)

    print(f'you chose {choices[my_choice]}')
    print(f'computer chose {choices[computer]}')

    # how do you choose who won and who lost? IF STATEMENT

    # rock
    # rock wins against scissors
    # choices = [rock, paper, scissors]
    # here we creates conditinal statements through indexing ---> choices[number] == choices[0]
    # choices[my_choice] == 'rock' is uncorrect, because our index doesn't retrieve a label ('rock'), like a string
    # it's variable (rock), so to make the conditional statement work we need to use the index again
    if choices[my_choice] == choices[0] and choices[computer] == choices[2]:
        print('you won')
    elif choices[my_choice] == choices[0] and choices[computer] == choices[1]:
        print('you lose')
    elif choices[my_choice] == choices[0] and choices[computer] == choices[0]:
        print('it\'s a tie')


    # paper
    # paper wins against rock
    if choices[my_choice] == choices[1] and choices[computer] == choices[0]:
        print('you won')
    elif choices[my_choice] == choices[1] and choices[computer] == choices[2]:
        print('you lose')
    elif choices[my_choice] == choices[1] and choices[computer] == choices[1]:
        print('it\'s a tie')

    # scissors
    # scissors win against paper
    if choices[my_choice] == choices[2] and choices[computer] == choices[1]:
        print('you won')
    elif choices[my_choice] == choices[2] and choices[computer] == choices[0]:
        print('you lose')
    elif choices[my_choice] == choices[2] and choices[computer] == choices[2]:
        print('it\'s a tie')
```
