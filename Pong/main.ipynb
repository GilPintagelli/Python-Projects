```python
from turtle import Turtle, Screen
import time

screen = Screen()

screen.setup(width=800,height=600)

screen.bgcolor("black")

screen.title("Pong")

screen.listen()

screen.tracer(0)

streak = int(screen.textinput("Shots","Choose between 10, 20 or 30 shots"))

pong = Turtle()


from paddle import Paddle

l_paddle = Paddle(-340,0)
r_paddle = Paddle(340,0)

screen.onkey(key="w",fun=l_paddle.go_up)
screen.onkey(key="s",fun=l_paddle.go_down)

screen.onkey(key="Up",fun=r_paddle.go_up)
screen.onkey(key="Down",fun=r_paddle.go_down)



from ball import Ball

ball = Ball()

from scoreboard import Score
game_score = Score()

game_on = True
while game_on:

    if game_score.l_score == streak or game_score.r_score == streak:
        game_on = False
        game_score.winner()
        game_score.game_over()
    else: 
        time.sleep(ball.ball_speed)

        # manually update the screen
        screen.update()
        ball.move()


        # if we change the ycor() our ball will bounce:
        # it is still on the x-axis but we only change the y-axis to change the ball's direciton
        if ball.ycor() > 280 or ball.ycor() < -280:
            # this bounce changes the y coordinate
            ball.bounce_y()     

        if ball.distance(r_paddle) < 50 and ball.xcor() > 300 or ball.distance(l_paddle) < 50 and ball.xcor() < -300:
            # this bounce is going to be different from the last, which changed the y coordinate
            # this one is going to to change both
            ball.bounce_x()
            # ball_speed-= 0.01        

        if ball.xcor() > 400:
            game_score.l_point()
            ball.reset_position()
            ball.move()
            ball.bounce_x()

        # left paddle misses
        if ball.xcor() < -400:
            game_score.r_point()
            ball.reset_position()
            ball.move()
            ball.bounce_x()

screen.exitonclick()
```
