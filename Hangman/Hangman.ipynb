```python
from hangman_art import logo
print(logo)
print('\n')

from hangman_art import stages
from hangman_words import word_list
import random

random_word = random.choice(word_list)

#for each letter in 'random_word', add a '_' to 'blanks'
blanks = [] # empty list
for letter in random_word:
    blanks+= '_'  # for each letter in random word add '_' to blanks
    
print(blanks)

last_letter = [] # create an empty list to gather the letters entered by the user
# instead of last_letter we can use blanks because if the letter is in the word then the letter is in blanks

end_game = False # we can easly change a boolean and still use lives separately   
# lives > 0  --> since the while loop is True, it keeps running even if the user wins

lives = 6
while not end_game: #--> condition --> here if we don't use 'not' it doesn't start looping becasue end_game is False

    print(stages[lives]) # use 'lives' as index to access 'stages'
    guess = input('Guess a letter: ').lower()

    
    if guess in blanks:
        print(f'you already used {guess}')
    
    # check if the letter is in the word
    for position in range(len(random_word)):  # this convert random_word in integers: position is a number
        letter = random_word[position] # use position as index to access random_word's letters
        if letter == guess:
            blanks[position] = guess
    
    # letter already guessed ---> if the user used a letter again they shouldn't lost a life, so we put it at the top
    if guess in last_letter:
        print(f'{guess} has already been used')

    elif guess not in random_word:
        lives-=1
        print(f'{guess} is not in the word')
       
        # we can put this if statement right into the main if statement or put it outside the main if statement
        # it works either way
        if lives == 0:
            end_game = True
            print('you lose')
            print(f'sorry, the correct word is {random_word}')     
    print(blanks)
    
    if '_' not in blanks:
        end_game = True # lives < 0 doesn't work because if user still has lifes left they can still play 
        print('you win')
        
    # step 3
    for letter in guess:
        last_letter.append(letter)
```
