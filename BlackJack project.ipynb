```python
import random

# ------------ FUNCTIONS---------------

def deal_card():
    """this function uses the random module and 
    takes a list and returns a random element
    from that list"""

    # Use the following list as the deck of cards:
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

    # the choice method is useful here --> it picks a card randomly
    random_card = random.choice(cards)
    return random_card

def calculate_score(list_cards):
    """this function takes a list of cards 
    and return the sum"""

    if sum(list_cards) == 21 and len(list_cards) >= 2 :
        return 0

    if sum(list_cards) > 21 and 11 in list_cards:
        list_cards.remove(11)
        list_cards.append(1)

    # score = sum(cards)
    return sum(list_cards)

def compare(user, comp):
    if user == comp:
        return "it's a draw"
    elif user > 21 or comp == 0:
        return "computer wins"
    elif user == 0 or comp > 21:
        return "user wins"
    elif user > comp:
        return "user wins"
    else:
        return "computer wins"
    
#---------------------------------------

def play_game():

    user_hand = []
    comp_hand = []


    # add cards to lists
    for n in range(2):
        user_hand.append(deal_card())
        comp_hand.append(deal_card())

    # this is called a FLAG
    end_game = False

    while not end_game:

        user_score = calculate_score(user_hand)
        comp_score = calculate_score(comp_hand)

        print(f'your card: {user_hand}, current score: {user_score}')
        print(f'computer\'s first card: {comp_hand[0]} and score is {comp_score}')

        if user_score == 0 or comp_score == 0 or user_score > 21 or comp_score > 21:
            end_game = True

        else:
            another_card = input("Do you want to draw another card? 'y' or 'n' ")
            if another_card == 'y':
                user_hand.append(deal_card())

            else:
                end_game = True

    while comp_score != 0 and comp_score < 17:
        comp_hand.append(deal_card())
        comp_score = calculate_score(comp_hand)

    print(f"Your final cards are {user_hand} and score is {user_score}, computer's score is {comp_score}")
    winner = compare(user=user_score, comp=comp_score)
    print(winner)
```

```python
while input("Do you want to play a game of BlackJack? type 'y' or 'n': ") == 'y':
    play_game()
```
